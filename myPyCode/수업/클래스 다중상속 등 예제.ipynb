```python
'''
1. 부모 클래스로 power를 추가한다 : 다중 상속 구현
    - powerOn(), powerOff()
2. 부모 클래스는 추상 클래스로 구현한다.
3. 전원을 켰을 때만 볼륨조정을 할 수 있게 한다.(조건문) whileTrue
4. 볼륨 조절은 max 값을 100으로 하고 min 값을 0으로 한다.(조건문) ,break while True
'''

1. tv
2. Radio
3. Speaker
4. 종료

메뉴: 1
1. 전원
2. 볼륨조정
3. 나가기

메뉴: 2
볼륨 레벨: 20
현재 TV의 볼륨은 X입니다.

```

```python
import abc

class Volume:
    vol = 0
    powerOn = True
    
    def volumeup(self, vol):
        self.vol += vol
        print("볼륨의 크기는 %d는 증가하여 현재 볼륨은 %d 입니다."%(vol, self.vol))
        
    def volumedown(self, vol):
        self.vol -= vol
        print("볼륨의 크기는 %d는 감소하여 현재 볼륨은 %d 입니다."%(vol, self.vol))

# 추상 메서드
class Power(metaclass=abc.ABCMeta):
    powerOn = True
    
    @abc.abstractclassmethod
    def on(self, powerOn):
        pass
    
    @abc.abstractclassmethod
    def off(self, powerOn):
        pass
    
####################################################################################################

class TV(Volume):
    
    def volumeup(self, vol):
        self.vol += vol
        print("TV 볼륨의 크기는 %d는 증가하여 현재 볼륨은 %d 입니다."%(vol, self.vol))
        
    def volumedown(self, vol):
        self.vol -= vol
        print("TV 볼륨의 크기는 %d는 감소하여 현재 볼륨은 %d 입니다."%(vol, self.vol))
        
    #on-off 구현(추상매서드)
    def on(self):
        self.powerOn 
        print("전원이 켜집니다.")
    
    def off(self):
        self.powerOn 
        print("전원이 꺼집니다.")
        
#####################################################################################################

class Radio(Volume):
    
    def volumeup(self, vol):
        self.vol += vol
        print("Radio 볼륨의 크기는 %d는 증가하여 현재 볼륨은 %d 입니다."%(vol, self.vol))
        
    def volumedown(self, vol):
        self.vol -= vol
        print("Radio 볼륨의 크기는 %d는 감소하여 현재 볼륨은 %d 입니다."%(vol, self.vol))
    
    #on-off 구현(추상매서드)
    def on(self):
        self.powerOn 
        print("전원이 켜집니다.")
    
    def off(self):
        self.powerOn 
        print("전원이 꺼집니다.")
        
#####################################################################################################

class Speaker(Volume):
    
    def volumeup(self, vol):
        self.vol += vol
        print("Speaker 볼륨의 크기는 %d는 증가하여 현재 볼륨은 %d 입니다."%(vol, self.vol))
        
    def volumedown(self, vol):
        self.vol -= vol
        print("Speaker 볼륨의 크기는 %d는 감소하여 현재 볼륨은 %d 입니다."%(vol, self.vol))
    
    #on-off 구현(추상매서드)
    def on(self):
        self.powerOn 
        print("전원이 켜집니다.")
    
    def off(self):
        self.powerOn 
        print("전원이 꺼집니다.")
```
