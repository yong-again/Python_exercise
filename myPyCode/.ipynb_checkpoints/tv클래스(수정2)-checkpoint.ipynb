```python
'''
1. 부모 클래스로 power를 추가한다 : 다중 상속 구현
    - powerOn(), powerOff()
2. 부모 클래스는 추상 클래스로 구현한다.
3. 전원을 켰을 때만 볼륨조정을 할 수 있게 한다.(조건문) whileTrue
4. 볼륨 조절은 max 값을 100으로 하고 min 값을 0으로 한다.(조건문) ,break while True
'''

1. tv
2. Radio
3. Speaker
4. 종료

메뉴: 1
1. 전원
2. 볼륨조정
3. 나가기

메뉴: 2
볼륨 레벨: 20
현재 TV의 볼륨은 X입니다.
```

```python
import abc

class Volume:
    vol = 0
    powerOn = True
    
    def volumeup(self, vol):
        self.vol += vol
        print("볼륨의 크기는 %d는 증가하여 현재 볼륨은 %d 입니다."%(vol, self.vol))
        
    def volumedown(self, vol):
        self.vol -= vol
        print("볼륨의 크기는 %d는 감소하여 현재 볼륨은 %d 입니다."%(vol, self.vol))
                
####################################################################################################

# 전원
class Power(metaclass=abc.ABCMeta):
    powerOn = True
    
    @abc.abstractclassmethod
    def on(self, powerOn):
        pass
    
    @abc.abstractclassmethod
    def off(self, powerOn):
        pass
    
####################################################################################################

class TV(Volume):
    #on-off 구현(다중상속)
    
    def On(self):
        self.powerOn 
        print("전원이 켜집니다.")
    
    def Off(self):
        self.powerOn 
        print("전원이 꺼집니다.")
        
    def volumeup(self, vol):
        self.vol += vol
        if self.vol > 100:
            print("볼륨은 100을 넘을 수 없습니다.")
            self.vol = 100
            
        else:
            print("TV 볼륨의 크기는 %d는 증가하여 현재 볼륨은 %d 입니다."%(vol, self.vol))
            
    def volumedown(self, vol):
        self.vol -= vol
        
        if self.vol < 0:
            print("볼륨은 0미만일 수 없습니다.")
            self.vol = 0
            
        else:
            print("TV 볼륨의 크기는 %d는 감소하여 현재 볼륨은 %d 입니다."%(vol, self.vol))
tv = TV()  
while True:
    
    print("\n1.TV\n2.Radio\n3.Speaker\n4. 종료")

    c1 = int(input("기기를 선택하세요: "))

    if c1 == 1:
        print("TV를 선택하였습니다.")
        print("\n1. 전원\n2. 볼륨 조정\n3. 나가기")

        while True:
            m1 = int(input("실행할 메뉴를 고르세요: "))

            if m1 == 1:
                print("TV 전원의 상태: {}".format(tv.On()))

                if tv.On:
                      print("전원이 켜져있습니다.")
                      
                else:
                    print("전원이 꺼져있습니다.")
                    retry = input("TV전원을 다시 키시겠습니까? (Y/N): ")

                    if retry == "Y":
                        tv.On()
                        continue

                    else:
                        continue

            elif m1 == 2:
                print("볼륨조정")

                if tv.On:
                    print("현재 볼륨은 {} 입니다.".format(tv.vol))
                    setvolume = int(input("TV볼륨 조정(1. 증가, 2. 감소): "))

                    if setvolume == 1:
                        set_up = int(input("증가할 볼륨의 크기: "))
                        tv.volumeup(set_up)

                    else: 
                        set_up = int(input("감소할 볼륨의 크기: "))
                        tv.volumedown(set_up)
                else:
                    print("전원을 먼저 키세요")
                        
                    
            elif m1 == 3:
                print("이전 메뉴로 돌아갑니다.")
                break

        else:
            print("잘못 입력하였습니다.")

    elif c1 == 2:
        print("Raido를 선택하였습니다.")

    elif c1 == 3:
        print("Speaker를 선택하였습니다.")

    elif c1 == 4:
        print("종료를 선택하였습니다.")
        break

    else:
        print("잘못 입력하였습니다.")
```
