```python
#가위바위보
import random as rd

#변수
com = rd.choice(['가위','바위','보'])
user = input("입력: ")

winner = 'none'

#if 조건문을 이용한 가위바위보
if com == '가위':
    if user == '가위':
        winner = 'none'
        
    elif user == '바위':
            winner = 'user'
        
    else:
        winner = 'computer'
        
if com == '바위':
    if user =='가위':
        winner = 'computer'
        
    elif user == '보':
        winner = 'user'
        
    else:
        winner = 'none'
        
if com == '보':
    if user == '가위':
        winner = 'user'
        
    elif user == '보':
        winner = 'none'
        
    else:
        winner = 'comupter'
        
if winner == 'computer':
    print("컴퓨터: ", com)
    print("사용자: ", user)
    print("승자: ","컴퓨터")
    
elif winner == 'user':
    print("컴퓨터: ", com)
    print("사용자: ", user)
    print("승자: ","사용자")
    
else:
    print("컴퓨터: ", com)
    print("사용자: ", user)
    print("승자: ", "무승부")
```

```python
# 계산기

res = 0

while True:
    cal = input("+,-,/,*,x 중 하나를 입력하시오: ")
    
    if cal == 'x': break
    
    num1 = int(input("첫번째 값을 입력하시오: "))
    num2 = int(input("두번째 값을 입력하시오: "))
    
    if cal == "+":
        res = num1+num2
    
    elif cal == "-":
        res = num1-num2
    elif cal == "/":
        res = num1/num2 
    elif cal == "*":
        res = num1*num2
    else:
        print("계산기를 종료합니다.")

print("결과: %d" % res)
```

```python
# 1~100 맞추기

import random as rd

print("1~100의 숫자는?: ")
answer = rd.randint(1, 100)
#print(answer)

guess = 0
count = 0

while guess != answer:
    print("당신의 답은?!")
    
    guess = int(input())
    count = count + 1

    if guess > answer:
        print("낮게!")
        
    elif guess < answer:
        print('높게!')
        
    else:
        continue
else:
    print("정답횟수:" + str(count))
```

```python
#가위바위보 승패출력
import random

win, lose= 0, 0
while True:
    com = random.choice(['가위','바위','보'])
    user = input("입력: ")
    winner= 'none'
    
    if user in ['x','q']: break
    
    if com == '가위':
        if user == '가위':
            winner = 'none'
            
        elif user == '바위':
            winner = 'user'
            win += 1
        else:
            winner = 'computer'
            
    if com == '바위':
        if user == '가위':
            winner = 'computer'
            lose +=1
        elif user == '보':
            winner = 'user'
            win += 1
        else:
            winner = 'none'
            
    if com == '보':
        if user == '가위':
            winner = 'user'
            win += 1
        elif user == '보':
            winner = 'none'
            
        else:
            winner = 'computer'    
            lose += 1
    if winner == 'computer':
        print("컴퓨터: ", com)
        print("사용자: ", user)
        print("승자: ","컴퓨터")
    
    elif winner == 'user':
        print("컴퓨터: ", com)
        print("사용자: ", user)
        print("승자: ","사용자")
    
    else:
        print("컴퓨터: ", com)
        print("사용자: ", user)
        print("승자: ", "무승부")
        
print("{}승 {}패".format(win, lose))
```

```python
# 짝수, 홀수의 합

num1, num2= 0, 0
while True:
    num = int(input(" "))
    
    if num == 0:
        break
    elif num % 2 == 0:
        num1 += num
    else:
        num2 += num
        
print("짝수의 합: ", num1, "홀수의 합: ", num2)
```
